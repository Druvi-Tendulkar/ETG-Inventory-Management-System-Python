## CASE 1

```python
print('********************************************************\n---------------Welcome to the Inventory-----------------\n********************************************************\n~~~~~~~~~~~~~~~~~~~~~~~~~~ Menu: ~~~~~~~~~~~~~~~~~~~~~~~\n1. View all the Items.\n2. Purchase Items.\n3. Add Items to the Inventory.\n4. Remove Item from Inventory.\n5. Exit.')
Choice = int(input("Pick Your Choice: "))

if(Choice == 1):
    for i in record:
        print(i,':',record[i])

    
#2. Purchase Items.    
if(Choice == 2):
    print('------------------Categories------------------\n1.Chocolate\n2.Electronics\n3.Makeup\n4.Stationery')
    category = input("Enter the Category Name of the item you want to purchase: ")
    
    for j in record:
            if(category == record[j]['Category']):
                print(j,'.',record[j]['Name'])
            #else:
                #print("Category Does not Exist")
            
    item_no = str(input("Enter the item Number you want to purchase: "))
    quantity = str(input("Enter the quantity: "))
                         
    if(int(quantity) > int(record[item_no]['Quantity'])):
        print("The number of items are less than the required quantity!")
        
    else:
        print('Item Name: ', record[item_no]['Name'])
        print('Item Price: ', record[item_no]['Price'])
        print('Billing Amount = ', int(record[item_no]['Price']) * int(quantity))
        record[item_no]['Quantity'] = str(int(record[item_no]['Quantity']) - int(quantity))
        
        js = json.dumps(record)              
        fd = open('IMS.json', 'w')
        fd.write(js)
        fd.close()
        
        #SALES
        sales={}
        transaction_id = len(sales)+1
        sales = {transaction_id :{'Name': record[item_no]['Name'], 'Price': record[item_no]['Price'], 'Quantity': quantity, 'Billing Amount': int(record[item_no]['Price']) * int(quantity)}}
        js = json.dumps(sales)              
        fd = open('Sales.json', 'a')
        fd.write(js)
        fd.close()       
                           
    
    
#3. Add Items to the Inventory.
elif(Choice == 3):
    print("1.Add Existing Item\n2.Add New Item\n")
    choice1 = int(input("Pick your Choice: "))
    if(choice1 == 1):
        for i in record:
            print(i,' ', record[i]['Name'])
        choice2 = int(input("Enter Item Number: "))
        print("Item: ", record[str(choice2)]['Name'])
        quantity = int(input("Enter Quantity: "))
        record[item_no] = {'Name': record[item_no]['Name'], 'Category': record[item_no]['Category'], 'Price': record[item_no]['Price'], 'Quantity': str( int(quantity) + int(record[item_no]['Quantity'])), 'Rating': record[item_no]['Rating']}
        print("Quantity Added!")
        js = json.dumps(record)
        fd = open('IMS.json', 'w')
        fd.write(js)
        fd.close() 
    elif(choice1 == 2):
        print("Item_ID: ", int(len(record)+1))
        Item = str(input("Enter Item Name: "))
        Category = str(input("Enter Category: "))
        Price = int(input("Enter Item  Price: "))
        Quantity = int(input("Enter Quantity: "))
        print("Rating: 0")
        Item_id = len(record)+1
        record[Item_id] = {Item_id: {'Name': str(Item), 'Category': str('Category'), 'Price': str('Price'), 'Quantity': str(int(Quantity) + int(record[item_no]['Quantity'])), 'Rating': '0'}}
        js = json.dumps(record)
        fd = open('IMS.json', 'w')
        fd.write(js)
        fd.close()
        print("Item added successfully!")
    
    
#4. Remove Item from Inventory.
elif(Choice == 4):
    item_id = str(input("Enter the Item ID you want to delete: "))
    for i in record.copy():
          if(item_id == i):
                del record[i]
                print("Item Deleted!")
                js = json.dumps(record)
                fd = open('IMS.json', 'w')
                fd.write(js)
                fd.close()
          
    
    
#5. Exit.
elif(Choice == 5):
    exit()

else:
    print("Choice not Present! Try Again.")
```
